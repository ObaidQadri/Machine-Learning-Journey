```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Data/canada_per_capita_income.csv")
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
per_capita_income = df['per capita income (US$)']
per_capita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
years_df = pd.read_csv("Data/years.csv")
years_df.head(5)
```

```python
p = reg.predict(years_df)
p
```

```python
years_df['predicted per capita income (US$)']=p
years_df.head(3)
```

```python
years_df.to_csv("Data/prediction_Assignment.csv")
```

```python
plt.xlabel('Year',fontsize =15)
plt.ylabel('Per capita income (US$)',fontsize =15)
plt.scatter(df.year,df['per capita income (US$)'], color="red", marker="+")
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
plt.show()
```
